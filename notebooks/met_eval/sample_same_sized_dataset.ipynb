### Load Denoised Data

```python
noise_list = [str(i) for i in range(0,60,10)]
noise_list
```

```python
import pandas as pd
denoised_base_pth = "..\\data\\sample\\train\\scaled\\denoised\\train_flipped_labels_"
#noised_base_pth = "..\\data\\sample\\train\\scaled\\train_flipped_labels_"
#test = pd.read_csv("..\\data\\sample\\test\\scaled\\test_scaled.csv")
#test_pth = "..\\data\\sample\\test\\scaled\\test_scaled.csv"
denoised_list =[]
noised_list=[]
for i in range(len(noise_list)):
    denoised_list.append(pd.read_csv(denoised_base_pth+noise_list[i]+".csv"))
    #noised_list.append(pd.read_csv(noised_base_pth+noise_list[i]+".csv"))
len(denoised_list)
```

### Down-size datasets with higher noise rates into noise rate 50% case

```python
sampled_denoised_base_pth = "..\\data\\sample\\train\\scaled\\denoised\\same_sized\\train_flipped_labels_"
for i in range(0,len(denoised_list)-1):
    df = denoised_list[i]
    label_1 = df.loc[df['label_flip'] == 1]
    label_0 = df.loc[df['label_flip'] == 0]
    filtered_1 = label_1.sample(n=7753)
    filtered_0 = label_0.sample(n=7753)
    filtered_df = pd.concat([filtered_1,filtered_0])
    #print(filtered_df.loc[filtered_df["label_flip"]==0])
    df_to_save = filtered_df.sample(frac=1).reset_index(drop=True)
    df_to_save.to_csv(sampled_denoised_base_pth+noise_list[i]+".csv",index=False)
   
```
